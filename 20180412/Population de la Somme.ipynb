# Obtenir le fichier et création d'une DataFrame

```python
import pandas as pd

xl = pd.read_excel('https://www.insee.fr/fr/statistiques/fichier/2525755/ensemble.xls', 
                   sheet_name='Communes',
                   skiprows=7)
```

# On affiche quelques lignes d'exemples

```python
xl.head()
```

# On ne garde que les communes dont le département est 80

```python
xl = xl[xl['Code département'] == '80']
xl.head()
```

# On trie par population, de manière décroissante

```python
xl = xl.sort_values(by=['Population municipale'], ascending=False)
xl.head()
```

# On ne garde que les 10 plus grandes villes

```python
big_cities = xl.head(10)
big_cities.head()
```

# Et on fait un joli graphique avec tout ça !

```python
%matplotlib inline
big_cities.plot(x='Nom de la commune',y='Population municipale',kind='bar')
```
