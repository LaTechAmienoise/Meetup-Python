# Manipulation d'images

```python
from matplotlib.pyplot import imshow
import numpy as np
import PIL.Image as pil

%matplotlib inline
```

```python
im = pil.open("earth.png")
tab = np.array(im)
imshow(tab)
```

```python
## Avec une matrice intermédiaire
```

```python
tab = np.array(im)
h,l,r = tab.shape #on sait que r=3
tab_output = np.zeros((h, l, 3), dtype = np.uint8)
for i in range(h):
    for j in range(l):
            tab_output[i][j] = tab[i][l-1-j]
imshow(tab_output)
```
